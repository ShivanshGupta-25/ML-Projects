```python
import os
import matplotlib.pyplot as plt
from PIL import Image 
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator,img_to_array,load_img
```

```python
# define the path to your image
image_path = "Datasets\Ironman_1.jpg"

# create a directory to save the augmented images
output_dir = 'Datasets\Images_Augmented'
os.makedirs(output_dir, exist_ok=True)
```

```python
# load the iamge and convert it to a Numpy array
img = load_img(image_path)
x = img_to_array(img)

# The ImageDataGenerator expects a 4D tensor : (batch_size, width, height, channels)
X = x.reshape((1,) + x.shape)
```

```python

# Initialise the ImageDataGenerator with augmentation parameters
datagen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = 'nearest'
)
```

```python
# Generate and save 10 augmented images
i = 0
for x in datagen.flow(X, batch_size=1, save_to_dir=output_dir,save_prefix='Ironman', save_format='png'):
    i += 1
    if i > 9:   # Generate 1o images
        break
print(f"Successfully generated and saved 10 images to {output_dir}")
```

```python
# define path to your Image 
image_path = 'Datasets\Ironman_1.jpg'

# Load the image and ensure it has an Alpha (transparancy) channel
pil_img_rgba = Image.open(image_path).convert('RGBA')

# convert the image ot a Numpy array
img_array = np.array(pil_img_rgba)
```

```python
# Extract each channel
red_channel = img_array[:,:,0]
green_channel = img_array[:,:,1]
blue_channel = img_array[:,:,2]
alpha_channel = img_array[:,:,3]

# create a figure to display all channels
plt.figure(figsize=(18,5))
```

```python
# plot the original image
plt.subplot(1,5,1)
plt.title('Original Image')
plt.imshow(pil_img_rgba)
plt.axis('off')

# Plot the Red channel
plt.subplot(1,5,2)
plt.title('Red Channel')
plt.imshow(red_channel, cmap='Reds')
plt.axis('off')

# Plot the green channel
plt.subplot(1,5,3)
plt.title('Green Channel')
plt.imshow(green_channel, cmap='Greens')
plt.axis('off')

# plot the blue channel
plt.subplot(1,5,4)
plt.title('Blue Channel')
plt.imshow(blue_channel, cmap='Blues')
plt.axis('off')

# plot the alpha channel
plt.subplot(1,5,5)
plt.title('Alpha Channel (Transparency)')
plt.imshow(alpha_channel, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
